```python
import pandas as pd
import matplotlib.pyplot as plt

df_precio_euro_m2 = pd.read_excel("./../datos/Precio euro_m2.xlsx")
df_precio_euro_m2 = df_precio_euro_m2.set_index('Lugar')

media = df_precio_euro_m2.mean(axis=0)

# Crear una lista con los años
years = df_precio_euro_m2.columns.astype(int)

# Crear una lista con los valores de la media
media_values = media.values
# Crear el gráfico de línea
plt.plot(years, media_values)

# Configurar el título y los ejes
plt.title('Media del Precio Euro/m2')
plt.xlabel('Año')

# Configurar los ticks del eje x
plt.xticks(years, rotation=90)


# Agregar las líneas de la cuadrícula
plt.grid(True)

# Mostrar el gráfico
plt.ylim(0)
plt.show()



```
