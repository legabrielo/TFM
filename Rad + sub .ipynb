## Merge DF Final


```python
# Para la manipulación de datos 

import pandas as pd
import numpy as np 
```

```python
# importamos df 

df = pd.read_excel(r"df-final.xlsx")
df
```

```python
# importamos el df de radiacion 

rad_df = pd.read_excel(r'Radiacion.xlsx')
rad_df
```

```python
# vemos si hay nas

df.info()
rad_df.info()
```

```python
df.Provincia = df.Provincia.str.replace('\n',"", regex=True)
df.Provincia.value_counts()
```

```python
# Cambiamos el tipo a fecha 
rad_df['Fecha'] = rad_df['Fecha'].str.replace("_", "-", regex=True)
rad_df
```

```python
# Hacemos un transpose de las columnas 
# Y ponemos las nuevas columnas
rad_df = rad_df.T.reset_index()
rad_df.head(10)
headers=rad_df.iloc[0]
rad_df.columns = headers
rad_df = rad_df.drop(rad_df.index[0])
rad_df=rad_df.rename(columns={'Fecha':'Provincia'})
rad_df
```

```python
# Hacemos regex 
rad_df.Provincia = rad_df.Provincia.str.replace('Jaen', 'Jaén', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Almeria', 'Almería', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Malaga', 'Málaga', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Cadiz', 'Cádiz', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Caceres', 'Cáceres', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Castellon', 'Castellón', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Avila', 'Ávila', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Leon', 'León', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Lerida', 'Lérida', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Cordoba', 'Córdoba', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Alava', 'Álava', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Lleida', 'Lérida', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Baleares', 'Islas Baleares', regex=True)
rad_df.Provincia = rad_df.Provincia.str.replace('Gipuzcoa', 'Guipuzcoa', regex=True)
```

```python
# Hacemos el merge
total_df = pd.merge(df, rad_df, how='left', on='Provincia')
total_df
```
