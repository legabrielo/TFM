```python
# Para la manipulación de datos
import pandas as pd
import numpy as np

# Para generar el modelo
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier

# Para guardar el modelo 
import pickle
```

```python
df=pd.read_csv("total_df.csv")
df=df.drop(columns="Unnamed: 0")
df
```

```python
df['rad-media'].fillna(100.0898789, inplace=True)
df
```

```python
df_suelos=pd.read_excel("Precio_medio_provincias.xlsx")
df_suelos
media_por_provincia = df_suelos.groupby('Lugar')['precio_medio'].mean().reset_index()
media_por_provincia.columns = ['Lugar', 'precio_medio_provincia']

# Fusionar los datos calculados con el DataFrame original
df_suelos = pd.merge(df_suelos, media_por_provincia, on='Lugar')
df_suelos=df_suelos[['Lugar','Anio','precio_medio','precio_medio_provincia']]
df_suelos
```

```python
ohe=OneHotEncoder()
feature_array=ohe.fit_transform(df[['Provincia', 'rad-media']]).toarray()
labels=ohe.categories_
np.array(labels,dtype='object').ravel()
```

```python
feature_label=np.array(labels,dtype='object').ravel()
```

```python
labelss = []
for i in feature_label:
    for j in i:
        labelss.append(j)
print(labelss)
```

```python
labels_str = [str(x) for x in labelss]
len(labels_str)
type(labels_str[77])
```

```python
features=pd.DataFrame(feature_array,columns=labels_str)

features.to_csv(r'C:\Users\Gabriel Castillo\Desktop\Cosas Gabriel\features.csv', index= False)
features
```

```python
todo=pd.concat([df, features ], axis = 1)
todo
```

```python
nneighbors = NearestNeighbors(n_neighbors= 52, metric = 'cosine').fit(features)
```

```python
df.iloc[10]
```

```python
features.iloc[[10]]
```

```python
liked=242
recom=250
# Find the indices of nearest neighbors for the liked restaurant
_, ind = nneighbors.kneighbors(features.iloc[[liked]])

# Display the liked location
print('Liked location ')
print('=' * 60)
print(df.iloc[liked, :])

print('\n')

# Display the recommended location
print('Recommended locations ')
print('=' * 60)
recommended_indices = ind[0][0:recom]  # Exclude the liked restaurant itself
recommended_locations = df.iloc[recommended_indices, :]
print(recommended_locations)
```

```python
recommended_locations
```

```python
presupuesto = 10000000  # 10 millones de euros

# Filtrar ubicaciones recomendadas dentro del presupuesto
filtered_locations = recommended_locations[recommended_locations['precio_medio_provincia'] <= presupuesto]

# Calcular la energía producida por metro cuadrado
filtered_locations['energia_por_metro_cuadrado'] = filtered_locations['rad-media'] * filtered_locations['area']

# Seleccionar la ubicación que produce la mayor energía dentro del presupuesto
ubicacion_optima = filtered_locations.loc[filtered_locations['energia_por_metro_cuadrado'].idxmax()]

print(ubicacion_optima)
```

```python
df_suelos.rename(columns={'Lugar': 'Provincia'}, inplace=True)
df_suelos
```

```python
extension=20000
presupuesto=40000000

# Calcular el costo total de adquirir la tierra para cada ubicación
df_suelos['costo_total'] = df_suelos['precio_medio_provincia'] * extension

# Filtrar ubicaciones dentro del presupuesto
ubicaciones_en_presupuesto = df_suelos[df_suelos['costo_total'] <= presupuesto]
concatenado = pd.merge(ubicaciones_en_presupuesto, recommended_locations, on='Provincia',how="inner")

# Calcular la producción de energía por metro cuadrado para cada ubicación en el presupuesto
concatenado['produccion_energia_por_metro_cuadrado'] = concatenado['rad-media'] * extension

# Seleccionar la ubicación con el mayor rendimiento de radiación dentro del presupuesto
ubicacion_optima = concatenado.loc[concatenado['produccion_energia_por_metro_cuadrado'].idxmax()]

ubicacion_optima
```
